```python
from sklearn.feature_extraction.text import CountVectorizer

cv_a = CountVectorizer(max_df = 0.95, stop_words = 'english')
#for i in range(100):
    
#    para_a_orig[i] = unicode(para_a_orig[i],'utf-8')
word_count_vec_a = cv_a.fit_transform(words_a)


list(cv_a.vocabulary_.keys())[:10]
print(word_count_vec_a.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)


tfidf_transformer.fit(word_count_vec_q)


#generate tf-idf for the given document
tf_idf_vector=tfidf_transformer.transform(word_count_vec_q)

```

```python
def sort_coo(coo_matrix):
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

sorted_items=sort_coo(tf_idf_vector.tocoo())


print(sorted_items[:10])
```

```python
def extract_topn_from_vector(feature_names, sorted_items, topn=10):
    """get the feature names and tf-idf score of top n items"""
    
    #use only topn items from vector
    sorted_items = sorted_items[:topn]

    score_vals = []
    feature_vals = []

    for idx, score in sorted_items:
        fname = feature_names[idx]
        
        #keep track of feature name and its corresponding score
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    #create a tuples of feature,score
    #results = zip(feature_vals,score_vals)
    results= {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]]=score_vals[idx]
    
    return results
```

```python
feature_names=cv_a.get_feature_names()

keywords=extract_topn_from_vector(feature_names,sorted_items,90000)


print(len(keywords))
```

```python
import unicodedata

keyw_list = []

print(len(keywords))

for i,j in keywords.items():
    #print(type(i),j)
    i = unicodedata.normalize('NFKD', i).encode('ascii','ignore') #unicode to string conversion
    #print(type(i))
    keyw = []
    #print(i,j)
    keyw.append(i.decode())
    keyw.append(j)
    keyw = tuple(keyw)
    keyw_list.append(keyw)

print(keyw_list[:10])
print(type(keyw_list))
```

```python
# Counting the word frequencies:
word_freq = nltk.FreqDist(itertools.chain(tokenized_text))
print ("Found %d unique words tokens." % len(word_freq.items()))

# Getting the most common words and build index_to_word and word_to_index vectors:
vocab = word_freq.most_common(vocabulary_size-1)
print(type(vocab))
```

```python
print(vocab[:10])
```

```python
final_vocab = []
F = 3144
for i in range(F):
    final_vocab.append(vocab[i])# append 200 words from author's vocab
for j in range(len(keyw_list)):
    final_vocab.append(keyw_list[j])# rest of the words are taken from the keyword list 

print(len(final_vocab))
#print(final_vocab[:1000])
```

```python
### Saving vocabulary:
print(final_vocab)
with open(vocabulary_file, 'wb') as v:
    pickle.dump(final_vocab, v)
```
